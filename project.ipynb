#### Delete the NAs of the dataset, if applicable (★★☆)

```python
import pandas as pd


# Lista de columnas con NaN que deseas reemplazar
columns_with_nans = ['rooms', 'bathrooms', 'surface', 'price']

# Imprime las columnas con NaN antes del reemplazo
print("Columnas con NaN antes del reemplazo:")
nan_columns_before = ds.columns[ds.isna().any()].tolist()
print(nan_columns_before)

# Reemplaza NaN con la mediana en las columnas numéricas relevantes
for column in columns_with_nans:
    median_value = ds[column].median()
    ds[column].fillna(median_value, inplace=True)

# Imprime las columnas con NaN después del reemplazo
print("\nColumnas con NaN después del reemplazo:")
nan_columns_after = ds.columns[ds.isna().any()].tolist()
print(nan_columns_after)



```

#### Calculate the sample mean and variance of the variables: price, rooms, surface area and bathrooms (★★★) 

```python
import pandas as pd
import numpy as np



# Calculate the sample mean and variance for each variable
mean_price = np.mean(subset_ds['price'])
variance_price = np.var(subset_ds['price'], ddof=1)  # ddof=1 for sample variance

mean_rooms = np.mean(subset_ds['rooms'])
variance_rooms = np.var(subset_ds['rooms'], ddof=1)

mean_surface = np.mean(subset_ds['surface'])
variance_surface = np.var(subset_ds['surface'], ddof=1)

mean_bathrooms = np.mean(subset_ds['bathrooms'])
variance_bathrooms = np.var(subset_ds['bathrooms'], ddof=1)

# Print the results
print(f"Mean Price: {mean_price:.2f}, Variance Price: {variance_price:.2f}")
print(f"Mean Rooms: {mean_rooms:.2f}, Variance Rooms: {variance_rooms:.2f}")
print(f"Mean Surface Area: {mean_surface:.2f}, Variance Surface Area: {variance_surface:.2f}")
print(f"Mean Bathrooms: {mean_bathrooms:.2f}, Variance Bathrooms: {variance_bathrooms:.2f}")
```

#### What is the most expensive home of each population? (★★☆) 

```python

import pandas as pd


# Subset the dataset as previously shown
populations_of_interest = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]
subset_ds = ds[ds['level5'].isin(populations_of_interest)]

# Function to get the row with the maximum price and its corresponding real estate name
def get_most_expensive_home(group):
    idx = group['price'].idxmax()
    return group.loc[idx]

# Group by population and find the most expensive home in each population along with real estate name
most_expensive_homes = subset_ds.groupby('level5').apply(get_most_expensive_home)

# Print the most expensive home in each population along with real estate name
print(most_expensive_homes[['level5', 'realEstate_name', 'price']])
```

#### Normalize the variable of prices for each population and plot the 4 histograms in the same plot (★★★) 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# List of populations of interest
populations_of_interest = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]

# Normalize the prices for each population
normalized_subset_ds = ds.copy()
for population in populations_of_interest:
    subset_population = normalized_subset_ds[normalized_subset_ds['level5'] == population]
    max_price = subset_population['price'].max()
    min_price = subset_population['price'].min()
    normalized_subset_ds.loc[subset_population.index, 'normalized_price'] = (subset_population['price'] - min_price) / (max_price - min_price)

# Create a single histogram plot for normalized prices using Seaborn
plt.figure(figsize=(10, 6))
for population in populations_of_interest:
    subset_population = normalized_subset_ds[normalized_subset_ds['level5'] == population]
    sns.histplot(data=subset_population, x='normalized_price', bins=20, kde=True, label=population)

plt.title('Normalized Prices Histogram for Different Populations')
plt.xlabel('Normalized Price')
plt.ylabel('Frequency')
plt.legend()
plt.show()



```

#### What can you say about the price per square meter between the towns of Getafe and Alcorcón?

```python
import pandas as pd



# Filter data for Getafe and Alcorcón
getafe_data = ds[ds['level5'] == 'Getafe']
alcorcon_data = ds[ds['level5'] == 'Alcorcón']

# Calculate average price per square meter for Getafe and Alcorcón
getafe_price_per_sqm = getafe_data['price'] / getafe_data['surface']
alcorcon_price_per_sqm = alcorcon_data['price'] / alcorcon_data['surface']

# Calculate average price per square meter for each town
avg_price_per_sqm_getafe = getafe_price_per_sqm.mean()
avg_price_per_sqm_alcorcon = alcorcon_price_per_sqm.mean()

# Print the results
print("Average Price per Square Meter in Getafe:", avg_price_per_sqm_getafe)
print("Average Price per Square Meter in Alcorcón:", avg_price_per_sqm_alcorcon)
```

the price per squaremeter is higher in Getafe!

#### Reload the data, is the average of "Valdemorillo" and "Galapagar" prices the same?

```python
import pandas as pd


# Filter data for Valdemorillo and Galapagar
valdemorillo_data = ds[ds['level5'] == 'Valdemorillo']
galapagar_data = ds[ds['level5'] == 'Galapagar']

# Calculate average price per square meter for Valdemorillo and Galapagar
valdemorillo_price_per_sqm = valdemorillo_data['price'] / valdemorillo_data['surface']
galapagar_price_per_sqm = galapagar_data['price'] / galapagar_data['surface']

# Calculate average price per square meter for each town
avg_price_per_sqm_valdemorillo = valdemorillo_price_per_sqm.mean()
avg_price_per_sqm_galapagar = galapagar_price_per_sqm.mean()

# Print the results
print("Average Price per Square Meter in Valdemorillo:", avg_price_per_sqm_valdemorillo)
print("Average Price per Square Meter in Galapagar:", avg_price_per_sqm_galapagar)

# Compare the averages
if avg_price_per_sqm_valdemorillo == avg_price_per_sqm_galapagar:
    print("The average price per square meter is the same in Valdemorillo and Galapagar.")
else:
    print("The average price per square meter is different in Valdemorillo and Galapagar.")
```

#### Make the same plot for 4 different populations (level5) and rearrange them on the same graph? (★★☆) 
Hint: make a scatter plot of each population using subplots.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# List of populations to compare
populations_to_compare = ['Fuenlabrada', 'Leganés', 'Getafe', 'Alcorcón']

# Create subplots
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
axes = axes.flatten()

# Create scatter plots for each population
for i, population in enumerate(populations_to_compare):
    subset_population = ds[ds['level5'] == population]
    sns.scatterplot(data=subset_population, x='surface', y='price', ax=axes[i])
    axes[i].set_title(f'Scatter Plot of Surface vs. Price - {population}')
    axes[i].set_xlabel('Surface Area')
    axes[i].set_ylabel('Price')

# Adjust layout
plt.tight_layout()
plt.show()
```
